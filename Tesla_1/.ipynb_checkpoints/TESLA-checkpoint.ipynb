<h4>Min</h4>

# Making a hashtable or dictionary of the closing data:

###  Analysis using hashtable: 
 - Easily allowed me to store data with average time complexity of O(1)
 - Allowed me to search data with time average complexity of O(1)
 
 - <span style="color:RED"> However, I still needed to sort the data once I have it, inorder for my histogram to be in increaing order.</span>.
### Alternate Approaches: 
 -  I could have used a pandas dataframe used df.groupby() and then df.count()
 
 
 


```python
# making a hashtable or dictionary of the closing data
closing = {}
for i in jsonstuuff['c']:
    intround = round(i)
    key =str(intround)
    if key in closing:
        closing[key]+=1 
    else:
        closing[key] = 1
```

```python
lists = sorted(closing.items())
x, y = zip(*lists)
```

```python
plt.figure(figsize=(20,10))
plt.bar(x,y)
plt.xlabel("Price($USD)",size=15)
plt.ylabel('Number of Occurrences',size=15)
plt.title("TSLA Stock 2019-2020(Closing Price for The Week Occurrences)",size =20)
```

<h1>Analysis of TSLA JAN (2019-Oct 2019) and (JAN 2020-Oct 2020)</h1>
<h2>Rate of change: </h2>
<h5> -I wanted to see how the two years compared </h5>

```python
lastyeardiff_distances = np.diff(jsonstuuff['c'])
thisyeardiff_distances = np.diff(thisyearstuuff['c'])
weeks = np.arange(33)
```

```python
# plt.plot(jsonstuuff['t'])
plt.figure(figsize=(20,10))
plt.plot(weeks[:32],lastyeardiff_distances[:32])
plt.plot(weeks[:32],thisyeardiff_distances[:32],'r--')
plt.xticks(rotation=45)
plt.xlabel("Week",size=15)
plt.ylabel('Change($USD/week)',size=15)
plt.title("TSLA Stock 2019 & 2020 Rate of Change($USD/week)", size =20)
plt.legend(['2019 closing High','2020 closing High'], loc='upper left')
```

```python
print("This Year average rate of change: "+ str(np.mean(thisyeardiff_distances)))
print("Last Year average rate of change: "+ str(np.mean(lastyeardiff_distances)))
```
